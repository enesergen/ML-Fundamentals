```python
#import pandas lib
import pandas as pd
#dataset url
url="https://raw.githubusercontent.com/tansutosun/MachineLearning/main/organics.csv";
#read dataset from github csv
df=pd.read_csv(url)
```

```python
#viewing dataset values with columns
df
```

```python
#check column names
df.keys()
```

```python
#Deleting ID Column
deleting_column_name='ID'
if deleting_column_name in df.keys():
  del df[deleting_column_name]
```

```python
#reviewing the current state
df.keys()
```

```python
#Check outliers data
df.isnull().sum()
```

```python
#Writing data type of  data frame columns
for data_type in df.keys():
  print(data_type+":"+str(df[data_type].dtypes))
```

```python
#Writing numeric column type
for column_name in df.keys():
  if pd.api.types.is_numeric_dtype(df[column_name]):
    print(column_name +" is numeric")
  else:
    print(column_name+" is not numeric")
```

```python
#Defining function for adjusting outliers data
def fillOutliersData(data_frame):
  #column key list
  column_list=data_frame.keys()

  for column_name in column_list:
    #if numeric, replace with median value for this column
    if pd.api.types.is_numeric_dtype(data_frame[column_name]):
      data_frame[column_name].fillna(data_frame[column_name].median(),inplace=True)
    #if not numeric, replace with mod value
    else:
      data_frame[column_name].fillna(data_frame[column_name].mode()[0],inplace=True)

  return data_frame
```

```python
#call function
df=fillOutliersData(df)

#check null values
df.isnull().sum()
```
