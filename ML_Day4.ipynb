````python
#Burada numeric ve kategorik dataları ayırıyoruz input dataları için
#Öncelikle kategorik olanları aldık
X_Categorical= df[['DemClusterGroup','DemGender']]
X_Categorical
````

````python
X_Numeric= df[['Dem Age','PromSpend']]
X_Numeric
````

````python
X_Input_Ctg=pd.get_dummies(X_Categorical)
X_Input_Ctg
````

````python
#Kategorik olanları numeric hale getirdik
X_Input=pd.merge(X_Numeric,X_Input_Ctg,left_index=True,right_index=True)
X_Input
````

````python
Y_Output=df['TargetBuy']
Y_Output
````

````python
#SKLearn kütüphanesi ile datamızı test ve train olarak bölüyoruz.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_Input, Y_Output, test_size=0.20, random_state=99)

````

````python
#Karar Ağacı yapılarının kullanımına geçiyoruz
#Decision tree classifier kullanıyoruz
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(criterion='gini',max_depth=2)
clf = clf.fit(X_train, y_train)
````


DecisionTreeClassifier fonksiyonunda criterion parametresi, karar ağacının düğümlerinin bölünmesi için kullanılan kriterin belirlenmesini sağlar. Bu parametre, karar ağacının hangi ölçüt üzerinden en iyi bölünmeyi yapacağını belirler.

criterion parametresi iki değer alabilir:

1-'gini': Bu değer, Gini impurity olarak adlandırılan bir ölçütü kullanarak düğümleri bölme kriterini belirtir. Gini impurity, bir düğümün homojenliğini ölçen bir metriktir. Bir düğümün Gini impurity değeri ne kadar düşük olursa, o düğüm o kadar homojen (yani tek bir sınıfa ait) olur.

2-'entropy': Bu değer, entropi (veya bilgi kazancı) olarak adlandırılan bir ölçütü kullanarak düğümleri bölme kriterini belirtir. Entropi, bir düğümün homojenliğini ölçen başka bir metriktir. Bir düğümün entropi değeri ne kadar düşük olursa, o düğüm o kadar homojen olur.

criterion parametresinin varsayılan değeri 'gini'dir. Genellikle gini veya entropy arasında seçim yapmak model performansı üzerinde farklı etkilere sahip olabilir. Bu nedenle, veri setine ve uygulamaya bağlı olarak hangi ölçütün kullanılacağı dikkatlice seçilmelidir.

````python
#şimdi ağacı ekrana bastıracağız
tree.plot_tree(clf)
````

````python
feat_importances = pd.Series(clf.feature_importances_, index=X_train.columns)
feat_importances.nlargest(13).plot(kind='barh')
````

````python
#Şimdi test datası ile modeli test edeceğiz.
y_pred = clf.predict(X_test)
````

````python
#Tahmin sonuçları yani y_pred ile y_test karşılaştıralım
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
````
