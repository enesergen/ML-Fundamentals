```python
#import pandas lib
import pandas as pd
#data csv url
url="https://raw.githubusercontent.com/tansutosun/MachineLearning/main/cs_accepts.csv"
#read dataset from github csv
df=pd.read_csv(url)
df
```

```python
df.isnull().any()
```

```python
#kendim yaptım
product_dummies=pd.get_dummies(df['PRODUCT'],dummy_na=True)
product_dummies
```

```python
import statsmodels.api as sm
y=df[['INCOME']]
y
```

```python
#girdi olarak kullanmak istediğimiz özellikler
x=df[['TMJOB1','AGE','CASH']]
x
```

```python
#sabit constant ekledik kolonlara ek olarak
x=sm.add_constant(x)
x
```

```python
#model eğitimi yapıldı
model=sm.OLS(y,x).fit()
```

```python
#modelin analiz sonucunu göstermek için yazıldı
model.summary()
```


Yukarıdaki sonuçlara göre age 	0.952 değeri ile etkisiz olduğunu anlıyoruz
çünkü katsayı değer aralığı -7.843,8.344 aralığındadır yani 0 da olabilir bu sayede denkleme etkisi olmaz
bu yüzden bu özelliği kullanmak çokta doğru değildir.



```python
predicted_df=pd.DataFrame(model.predict(x),columns=['predicted'])
predicted_df
```

```python
input_df=pd.DataFrame(x)
input_df
```

```python
output_df=pd.DataFrame(y)
output_df
```

```python
final_df=pd.concat([input_df,output_df,predicted_df],axis=1)
final_df
```

```python
final_df['error']=final_df['predicted']-final_df['INCOME']
final_df
```
